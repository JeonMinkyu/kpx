# Walk Forward
stacking model

#### yhat, y plot

```python
def y_plot(datetime, y1, y1_hat, isshow=False, file='walk_forward'):
    plt.figure(figsize=(20, 6))
    plt.plot(datetime, y1, label='y_test')
    plt.plot(datetime, y1_hat, label='y_test_hat')
    plt.legend()
    if isshow:
        plt.show()

    filename = './figure/'+file+'.png'
    plt.savefig(filename)
```

#### mse plot

```python
def mse_plot(mse, x_train, num_bins=1000, isshow=True):
    
    plt.figure(figsize=(20, 6))
    relative_density_plot(mse['index'], mse['mse'], num_bins=1000, show=False)
    plt.axvline(x=x_train.shape[0])
    plt.figure(figsize=(20, 6))
    plt.plot(mse['index'], mse['mse'], color = 'green')
    plt.axvline(x=x_train.shape[0])
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(prop={'size':15})
    if isshow:
        plt.show()
```

#### power moving average

```python
def power_ma(power, datetime, datetime_train isshow=True):
    
    plt.figure(figsize=(20,6))
    plt.plot(datetime, power, label='original',alpha=0.2)
    plt.plot(datetime, power.rolling(48).mean(), label='ma_48',alpha=0.7)
    plt.plot(datetime, power.rolling(150).mean(), label='ma_150',alpha=0.7)
    plt.plot(datetime, power.rolling(600).mean(), label='ma_600',alpha=0.7)
    plt.axvline(x=datetime_train.iloc[-1])
    plt.legend()
    if isshow:
        plt.show()
```

```python
split_nums = [30]
num = [10]
result = pd.DataFrame(columns=['train_r2', 'test_r2','train_mae','test_mae'])
train = []
test = []
train_mae = []
test_mae = []
feature_importance = []

for split_num in split_nums:
    for n in num:

        lst = list(range(0,len(X_train)))
        lst_2 = lst[:len(lst):len(lst)//split_num]

        split_lst = []
        for i in range(len(lst_2)-1):
            split_lst.append(lst[lst_2[i]:lst_2[i+1]])

        for i in range(len(lst_2)-(n-1)):
            train_index = range(lst_2[i], lst_2[i+n-2])
            test_index = range(lst_2[i+n-2], lst_2[i+n-1])

            x_train, x_test = X_train.iloc[train_index] , X_train.iloc[test_index]

            y_1 = y_train['Power Generation(kW)+0']
            y1_train, y1_test = y_1.iloc[train_index], y_1.iloc[test_index]

            y_2 = y_train['Power Generation(kW)+1']
            y2_train, y2_test = y_2.iloc[train_index], y_2.iloc[test_index]

            y_3 = y_train['Power Generation(kW)+2']
            y3_train, y3_test = y_3.iloc[train_index], y_3.iloc[test_index]

            print('TRAIN :', len(x_train), 'TEST :', len(x_test))
            datetime_train, datetime_test = df_train['datetime(forecast)'].iloc[train_index], df_train['datetime(forecast)'].iloc[test_index]
            model = StackedRegressor()

            model.fit(x_train, y1_train, y2_train, y3_train)
            y1_train_hat, y2_train_hat, y3_train_hat = model.predict(x_train)
            y1_test_hat, y2_test_hat, y3_test_hat = model.predict(x_test)

            y1_train_hat = pd.Series(y1_train_hat)
            y2_train_hat = pd.Series(y2_train_hat)
            y3_train_hat = pd.Series(y3_train_hat)

            y1_test_hat = pd.Series(y1_test_hat)
            y2_test_hat = pd.Series(y2_test_hat)
            y3_test_hat = pd.Series(y3_test_hat)

            y_train_hat = pd.concat([y1_train_hat, y2_train_hat, y3_train_hat], ignore_index=True)
            y_train = pd.concat([y1_train, y2_train, y3_train], ignore_index=True)

            y_test_hat = pd.concat([y1_test_hat, y2_test_hat, y3_test_hat], ignore_index=True)
            y_test = pd.concat([y1_test, y2_test, y3_test],ignore_index=True)
            
            
            #####train,test plot###
            y_plot(datetime_test, y1_test, y1_test_hat, isshow=False, file='walk_forward')

            train_r2_score = r2_score(y_train_hat, y_train)
            test_r2_score = r2_score(y_test_hat, y_test)
            print('train:',train_r2_score)
            print('test:',test_r2_score)

            nMAE_train = nMAE(y_train, y_train_hat)
            nMAE_test = nMAE(y_test, y_test_hat)

            ##### mse plot ####
            mse_train = (y_train_hat - y_train)**2
            mse_test = (y_test_hat - y_test)**2

            mse = pd.concat([mse_train, mse_test]).reset_index(drop=False)
            mse['index'] = list(range(mse.shape[0]))
            print('mse:',mse.columns)
            mse = mse.rename(columns={0:'mse'})
            mse_plot(mse, x_train)

            #### moving average plot ####
            power = pd.concat([y_train, y_test]).reset_index(drop=True)
            datetime = pd.concat([datetime_train, datetime_test]).reset_index(drop=True)
            
            power_ma(power, datetime, datetime_train, isshow=True)

            #### feature importance ####
#             plot_importance(model)
#             plt.show()

            result.loc[i] = [train_r2_score, test_r2_score, nMAE_train, nMAE_test]

    result.loc[i+1] = [result['train_r2'].mean(), result['test_r2'].mean() ,result['train_mae'].mean(), result['test_mae'].mean()]
    train.append(result['train_r2'].mean())
    test.append(result['test_r2'].mean())
    train_mae.append(result['train_mae'].mean())
    test_mae.append(result['test_mae'].mean())
    display(result)
#                 filename = './result/ma_feature_173_splitnum_'+str(split_num)+'n_'+str(n)+'result.csv'
#                 result.to_csv(filename)
            # print(result['test'].iloc[4])

print('train r2:',sum(train)/3)
print('test r2:',sum(test)/3)
print('train mae:',sum(train_mae)/3)
print('test mae:',sum(test_mae)/3)
```

# Nested Cross Validation
XGBoost

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import r2_score
from lightgbm import LGBMRegressor

def Nested_cv(n_split, X, y):
    train_y_hat_list = []
    test_y_hat_list = []
    y_train_list = []
    y_test_list = []
    
    tscv = TimeSeriesSplit(n_splits=n_split)
    TimeSeriesSplit(max_train_size=None, n_splits=n_split)
    result = pd.DataFrame(columns=['train', 'test'])
    for idx, (train_index, test_index) in enumerate(tscv.split(X)) :
        print('TRAIN:', train_index.shape[0], 'TEST:', test_index.shape[0])
        x_train, x_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        
        y_train_list.append(y_train)
        y_test_list.append(y_test)
        
        model = XGBRegressor()
        model.fit(x_train, y_train)
        y_train_hat = model.predict(x_train)
        y_test_hat = model.predict(x_test)
        
        train_y_hat_list.append(y_train_hat)
        test_y_hat_list.append(y_test_hat)
        
        plot_importance(model)
        plt.show()
        result.loc[idx] = [r2_score(y_train_hat, y_train), r2_score(y_test_hat, y_test)]
    #     for column in x_train.columns :
    #         relative_density_plot(x_train[column], y_train, label=[column, 'y'], num_bins=100, show=False)
    #         relative_density_plot(x_test[column], y_test, label=[column, 'y'], num_bins=100, show=False)
    #         plt.show()
    display(result)
    
    return result
```

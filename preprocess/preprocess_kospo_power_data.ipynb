## 풍력
- [링크](https://www.data.go.kr/dataset/15012779/fileData.do)
- [한국의 풍력발전소](https://snugis.tistory.com/47)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import datetime
import pickle
import math
```

```python
def first_aid(df) :
    df = df.drop(df.columns[[1, 2, 3, 4, 5, 6, 7]], axis=1)
    df.columns.values[0] = 'date'
    df['date'] = pd.to_datetime(df['date'])
    return df
```

```python
def create_df(df, location) :
    df_new = pd.DataFrame(columns=['date', 'datetime', 'location', '발전량(kW)'])
    df['date'] = pd.to_datetime(df['date'])
    
    for row_idx in range(df.shape[0]) :
        for column_idx in range(df.shape[1]-1) :
            
            df_new.loc[row_idx+column_idx] = [df.iloc[row_idx]['date'],
                                              datetime.datetime(df.iloc[row_idx]['date'].year,
                                                               df.iloc[row_idx]['date'].month,
                                                               df.iloc[row_idx]['date'].day,
                                                               column_idx,
                                                               0,
                                                               0),
                                              location,
                                              df.iloc[row_idx][column_idx+1]
                                             ]
    return df_new
```

## 성산풍력발전소
- 제주도 서귀포시 성산읍 수산리

```python
df_sungsan = pd.read_csv(os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'raw', 'kospo', 'wind', '한국남부발전_성산풍력발전실적_201712.csv')), encoding="cp949")
display(df_sungsan.head())
df_sungsan = first_aid(df_sungsan)
display(df_sungsan.head())
display(df_sungsan.tail())
location = 'Korea, Jeju-do, Seogwipo-si, Seongsan-eup'
df_sungsan = create_df(df_sungsan, location)
display(df_sungsan.head())
```

#### save

```python
df_kospo_a = df_sungsan.copy()
df_kospo_a.to_pickle(os.path.abspath(os.path.join(os.getcwd(), '..', 'data', 'df_kospo_a.pkl')))
```
